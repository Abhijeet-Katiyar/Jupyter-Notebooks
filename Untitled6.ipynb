```python
import pandas as pd
import matplotlib.pyplot as plt
Nifty_data=pd.read_csv("E:/summer/NIFTY50.csv",parse_dates=['Date'],index_col=['Date'])
data=Nifty_data['Close']
data.head()
```

# Testing stationarity

```python
def stationary_test(series):
# plotting rolling statistics    
    plt.plot(series,color='grey')
    plt.plot(series.rolling(window=252).mean(),label='yearly rolling mean',color='r')
    plt.plot(series.rolling(window=252).std(),label='yearly rolling std',color='black')
    plt.legend()
    plt.show()
# Adfuller test:    
    from statsmodels.tsa.stattools import adfuller
    test=adfuller(series,autolag='AIC')
    test_rslt=pd.Series(test[0:4],index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in test[4].items():
        test_rslt['Critical Value (%s)'%key] = value
    print(test_rslt)
    if(test[1]>0.05):
        print("Time Series is not stationary")
    else:
        print("Time series is stationary")
```

```python
stationary_test(data)
```

# Removing trend

### Moving avg

```python
mov_avg=data.rolling(window=21).mean()
data_mov_diff=data-mov_avg
data_mov_diff.dropna(inplace=True)
```

```python
stationary_test(data_mov_diff)
```

**test statistics is greater than 10% critical values so, time series is stationary with  90%  confidence**

```python
# exponential weighted moving average
exp_avg=pd.ewma(data,halflife=21)
data_ewma_diff=data-exp_avg
```

```python
stationary_test(data_ewma_diff)
```

**test sattistics less than 1% critical values so,we can say time series is stationary with more than 99% confidence**

### differencing

```python
data_diff=data-data.shift()
data_diff.dropna(inplace=True)
```

```python
stationary_test(data_diff)
```

**test statistics is greater than 10% critical values so, time series is stationary with  90%  confidence**

### Decomposing

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(data,freq=252)
```

```python
plt.plot(decomposition.trend)
plt.show()
plt.plot(decomposition.seasonal)
plt.show()
plt.plot(decomposition.resid)
plt.show()
```

```python
decompo_resid=decomposition.resid
decompo_resid.dropna(inplace=True)
stationary_test(decompo_resid)
```

**test sattistics less than 1% critical values so,we can say time series is stationary with more than 99% confidence**

# forecasting

### ACF and PACF plotting

```python
from statsmodels.tsa.stattools import acf, pacf
import numpy as np
```

```python
lag_acf = acf(data_diff, nlags=5)
lag_pacf = pacf(data_diff, nlags=5, method='ols')
```

```python
# plot ACF
plt.plot(lag_acf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(data_diff)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(data_diff)),linestyle='--',color='gray')
plt.title('Autocorrelation Function')
plt.show()
```

```python
# plot PACF
plt.plot(lag_pacf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(data_diff)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(data_diff)),linestyle='--',color='gray')
plt.title('Partial Autocorrelation Function')
plt.show()
```
